```python
import numpy as np
```

```python
col = 20
row = 100

mseg = 100*np.random.randint(10,100, size=(1, col))
mseg = mseg.tolist()
mseg = mseg[0]
print(mseg)
```

```python
rutil = 4*np.ones(col)
rutil = rutil.tolist()
print(rutil)
```

```python
fixedcost = 100*np.random.randint(50,250, size=(1, row))
fixedcost = fixedcost.tolist()
fixedcost = fixedcost[0]
print(fixedcost)
```

```python
profit = np.random.randint(20,100, size=(1, row))
profit = profit.tolist()
profit = profit[0]
print(profit)
```

```python
utility = np.random.randint(-5,row, size=(row, col))
utility = [list(u) for u in utility]
print(utility)
```

```python
#Write your file here
f = open("dataset.txt","w")
f.write(str(mseg))
f.write("\n\n")
f.write(str(utility))
f.write("\n\n")
f.write(str(fixedcost))
f.write("\n\n")
f.write(str(profit))
f.write("\n\n")
f.write(str(rutil))
f.close()
```
